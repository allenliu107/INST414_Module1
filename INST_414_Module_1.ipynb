```python
# prompt: change pay so that it's a min column and max column if there is only one number just use that both as max and min. if any values are per hour remove them

import re

def clean_pay(pay_list):
    cleaned_pay = []
    for pay_str in pay_list:
        # Remove "per hour" or similar phrases
        pay_str = re.sub(r'\b(per hour|hourly|an hour)\b', '', pay_str, flags=re.IGNORECASE)

        # Extract numbers and dollar signs
        numbers = re.findall(r'(\d+[\.,]?\d*)', pay_str)
        if numbers:
          cleaned_numbers = []
          for num_str in numbers:
            try:
              num = float(num_str.replace(',', ''))
              cleaned_numbers.append(num)
            except ValueError:
              pass  # Ignore invalid number strings
          if cleaned_numbers:
            cleaned_pay.extend(cleaned_numbers)

    if not cleaned_pay:
      return [np.nan, np.nan]

    if len(cleaned_pay) == 1:
        return [cleaned_pay[0], cleaned_pay[0]]  # Use the same value for min and max

    return [min(cleaned_pay), max(cleaned_pay)]

df_pay[['min_pay', 'max_pay']] = pd.DataFrame(df_pay['Pay'].apply(clean_pay).tolist(), index= df_pay.index)
df_pay = df_pay.drop('Pay', axis=1)
display(df_pay.head(100))
```
